```python
from bs4 import BeautifulSoup
```

```python
import requests
```

```python
url="https://www.flipkart.com/search?q=biography%20books&otracker=search&otracker1=search&marketplace=FLIPKART&as-show=on&as=off"
```

```python
allow=requests.get(url)
```

```python
allow
```

```python
soup=BeautifulSoup(allow.text)
```

```python
soup
```

```python
title=soup.select(".s1Q9rs")
```

```python
title
```

```python
title_list=[]
```

```python
for a in title:
    title_list.append(a.text)
```

```python
title_list
```

```python
rating
```

```python
rating_list=[]
```

```python
rating_list
```
